```python
#Create a sample 2D NumPy array
sample_array = np.array([[10, 11, 12],[13, 14, 15],[16, 17, 18]])

print("Original array:")
print(sample_array)
print("-" * 20)

#Demonstrate accessing individual elements using integer indexing
print("Accessing element at row 0, column 0:", sample_array[0, 0])
print("Accessing element at row 1, column 2:", sample_array[1, 2])
print("-" * 20)

#Demonstrate accessing rows and columns using slicing
print("Accessing the first row:", sample_array[0, :])
print("Accessing the second column:", sample_array[:, 1])
print("Accessing the first two rows and the last two columns:")
print(sample_array[0:2, 1:3])
print("-" * 20)

#Demonstrate modifying array elements by assigning new values
sample_array[0, 0] = 99
print("Array after modifying element at (0, 0):")
print(sample_array)
print("-" * 20)

#Demonstrate modifying a slice by assigning new values
sample_array[1:3, 1:3] = [[50, 51], [52, 53]]
print("Array after modifying a slice (rows 1-2, columns 1-2):")
print(sample_array)
```

```python
"""
PROJECT: Simulate and Analyze Image Data
- Images are just 3D NumPy arrays (height, width, RGB channels)
- Practice array manipulation like real AI image tasks
"""

import numpy as np
import matplotlib.pyplot as plt

# Simulate a 100x100 RGB image (values 0-255)
# Shape: (height, width, channels)
image = np.random.randint(0, 256, size=(100, 100, 3), dtype=np.uint8)

print(f"Image shape: {image.shape}")
print(f"Image size: {image.size} pixels")

# Extract individual channels
red_channel = image[:, :, 0]
green_channel = image[:, :, 1]
blue_channel = image[:, :, 2]

# Image statistics
print(f"\nRed channel - Mean: {red_channel.mean():.2f}, Std: {red_channel.std():.2f}")
print(f"Green channel - Mean: {green_channel.mean():.2f}, Std: {green_channel.std():.2f}")
print(f"Blue channel - Mean: {blue_channel.mean():.2f}, Std: {blue_channel.std():.2f}")

# Convert to grayscale (common preprocessing step)
# Standard formula: 0.299*R + 0.587*G + 0.114*B
grayscale = 0.299 * red_channel + 0.587 * green_channel + 0.114 * blue_channel
grayscale = grayscale.astype(np.uint8)

print(f"Grayscale shape: {grayscale.shape}")

# Visualize
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].imshow(image)
axes[0].set_title("Original RGB Image")
axes[0].axis('off')

axes[1].imshow(grayscale, cmap='gray')
axes[1].set_title("Grayscale Image")
axes[1].axis('off')

plt.tight_layout()
plt.show()

# Image transformations
flipped = np.flip(image, axis=0)  # Flip vertically
cropped = image[25:75, 25:75]      # Crop center 50x50
brightened = np.clip(image + 50, 0, 255).astype(np.uint8)  # Increase brightness
```
